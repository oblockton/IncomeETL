```python
import pandas as pd
import sqlalchemy
```

```python
# Housing Price Index CSV path
csv_path = "Resources/HPI_PO_state.xls"
```

```python
# For XLS use read_excel
pd_read = pd.read_excel(csv_path, sheetname ="Downloadable--State (pls cpy)")
```

```python
pd_read.head()
```

```python
# Create a copy for transformation.
hpi_state = pd_read.copy()
```

```python
hpi_state.head()
```

```python
# Drop Warning column
hpi_clean = hpi_state.drop(columns=['Warning'])
```

```python
hpi_clean.head()
```

```python
# The table containing data on regional income provides date/time information in a string that combines year and quarter.
# In this table year and quater are seperate fields. This function creates a new row with the correct/matching date format.
def quarterizer(row):
    year = row["yr"]
    quarter = row["qtr"]
    str_rep = str(year) + "Q" + str(quarter)
    return str_rep
hpi_clean["TimePeriod"] =hpi_clean.apply(quarterizer, axis=1)
```

```python
hpi_clean.head()
```

```python
# create a engine.mysql connector
engine = sqlalchemy.create_engine('mysql://root:heata4321.@localhost:3306') # connect to server
 # select new db
# use the new db
# continue with your work...
```

```python
# Point to databse to use. Database previously created through Workbench
engine.execute('USE hpi_db')
```

```python
# Test table existed
engine.execute('drop table hpi_db')
```

```python
# Construct Sql table/class from the housing index dataframe.
hpi_clean.to_sql('hpi_table', engine, schema=None, if_exists='replace')
```

```python
# store query result
results = engine.execute('SELECT * from hpi_table')
```

```python
# display query results.
for result in results:
    print(result)
```
